# Funciones algebraicas

### Funci贸n lineal

Tiene la forma de $$f(x)=mx + b$$ donde $m$ y $b$  $\in R$. 

$m$ puede ser calculada por: $$m=\frac{y_{2}-y_{1}}{x_{2}-x_{1}}$$

y $b$ es el punto de corte con el eje $y$. Su dominio es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = (-\infty, \infty)$




```python
N = 100   #define number of points
m = -1    #define slope
b = 3     #define cutting point for the equation

def f(x):
    return m*x+b

x = np.linspace(-10,10, num=N)    #define linespace starting from -10 to 10 dividen in N(defined above) 
                                  #parts

y =f(x)

fig, ax = plt.subplots()
ax.plot(x,y)
ax.grid()


```

### Funciones polin贸micas

Tiene la forma de $$P(x)=a_{n}x^{n} + a_{n-1}x^{n-1}+...+a_{2}x^{2}+a_{1}x + a_{1}$$

a una funci贸n que tiene esta forma se le llama polinomio de grado $n$. A los elementos $a$ los llamaremos coeficientes donde $a \in R$. 

**Por ejemplo:**

$$P(x)= 2x^{7} - x^{4} + 3x^{2} + 4$$

que es un polinomio de grado 7.


### Funciones potencia

Hay unas funciones que son un caso particular de las funciones polin贸micas que son las funciones potencia, las cuales tienen la forma:


$$f(x)= x^{a}, a \in R$$ 


**Por ejemplo:**

$$f(x)= x^{2}$$

El dominio de $f(x)=x^{2}$ es $Dom_{f} = (-\infty, \infty)$. Su imagen es $Im_{f} = [0, \infty)$





```python
def f(x):
    return (2*x**7)-(x**4)+(3*x**2)+4

y=f(x)

plt.plot(x,y)
plt.grid()
```

### Funci贸n exponencial

Tienen la forma de $$f(x)=a^x$$ donde la base $a$ es una constante positiva. Un gran ejemplo de una funci贸n exponencial es usando la base como el n煤mero de euler:

```python
def f(x):
    return np.exp(x)     #euler number in power of x using numpy

y = f(x)
plt.plot(x,y)
```

### Funci贸n logaritmo

El logaritmo est谩 definido por la **relaci贸n**:

$$log_{b}(x) = n \Longleftrightarrow x=b^n$$ 

donde: 



*   $b$ es la base.
*   $n$ es el exponente al que est谩 elevado la base.
*   $x$ es el resultado de elevar la base $b$ al exponente $n$

**Ejemplo:**

Teniendo b=2 y n=8, entonces:

$$2^8=256$$

Por lo que $x=256$. Calculando el logaritmo base 2 de $x$ es:

$$log_{2}(256) = 8$$





```python
def f(x):
    return np.log2(x)     #natural logarithm of x

x = np.linspace(0.001,256, num=1000)    #define linespace starting from 0.1 to 10
y = f(x)
plt.plot(x,y)
```

# Funci贸n seccionada

Son funciones que tienen diferentes valores definidos por un intervalo. Por ejemplo la funci贸n escal贸n de Heaviside: 

$$H(x) = 
     \begin{cases}
        0, &\quad \text{para, } x < 0 \\
        1,  &\quad\text{para. } x \ge 0 \\
     \end{cases}
$$

```python
def H(x):
    return (x >= 0).astype(int)

N = 1000
x = np.linspace(-10,10, num=N)
y = H(x)
plt.plot(x,y)
```
