# Medidas de tendencia central

### Media = $$ \frac{1}{N} \sum_{i=1}^N x_i  $$ 

### Mediana(impar) = $$ x_{(n+1)/2}^{\text{ordered}} $$

### Mediana(par) = $$ \frac{x_{n/2}^{\text{ordered}} + x_{n/2+1}^{\text{ordered}}}{2} $$

### Moda = $$ x_k $$ donde $$ \text{Freq}(x_k) = \max{(\text{Freq}(x_i))} $$

* **pro tip:** usar seaborn: https://seaborn.pydata.org/tutorial/distributions.html

```python
import seaborn as sns
```

```python
sns.displot(df,x = 'price_usd', hue = 'manufacturer_name')
"""
This code creates a distribution plot using the seaborn library.

Parameters:
- df: DataFrame
    The DataFrame containing the data to be plotted.
- x: str
    The name of the column in the DataFrame to be plotted on the x-axis. In this case, it is 'price_usd'.
- hue: str
    The name of the column in the DataFrame to be used for color encoding. In this case, it is 'manufacturer_name'.

Description:
The `sns.displot` function is used to create a distribution plot of the 'price_usd' column from the DataFrame `df`.
The `hue` parameter is set to 'manufacturer_name', which means that the plot will use different colors to represent
different manufacturers. This allows for a visual comparison of the distribution of 'price_usd' across different
manufacturers.
"""
```

```python
sns.displot(df, x = 'price_usd', hue = 'engine_type')
```

```python
sns.displot(df, x = 'price_usd', hue = 'engine_type', multiple = 'stack')
```

```python
Q7_df = df[(df['manufacturer_name'] == 'Audi') & (df['model_name'] == 'Q7')]
sns.histplot(Q7_df, x='price_usd', hue= 'year_produced')
```
